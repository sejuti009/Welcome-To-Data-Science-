```python
import matplotlib.pyplot as plt
```

```python
students = ["Sanjay", "Rahul", "Karan", "Wasin", "Ron", "Ajay", "John", "Emma"]
```

```python
students_marks = [35,50, 20, 45, 25, 48, 25, 40]
```

```python
marks_perc = []
```

```python
for x in students_marks:
   res = (x/50)*100
   marks_perc.append(res)
print(marks_perc)
```

```python
def line_chart():
  plt.plot(students,students_marks)
  plt.title("Students Marks Graph")
  plt.xlabel("Students Names")
  plt.ylabel("Students Marks")
  plt.show()
line_chart()
```

```python
#Bar Chart
def perc_bar_chart():
  plt.bar(students,students_marks)
  plt.title("Students Percentage Graph")
  plt.xlabel("Students Names")
  plt.ylabel("Students Marks")
  plt.show()
perc_bar_chart()
```
