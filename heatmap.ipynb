```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Example dataset (students' marks in subjects)
df = pd.DataFrame({
    "Math": [80, 90, 70, 60, 85],
    "Physics": [75, 88, 69, 65, 82],
    "Chemistry": [78, 85, 72, 62, 88],
    "English": [60, 70, 80, 75, 65],
    "Biology": [82, 87, 77, 64, 80]
}, index=["Student1", "Student2", "Student3", "Student4", "Student5"])

# Plot heatmap of raw values
plt.figure(figsize=(8, 6))
sns.heatmap(df, annot=True, cmap="YlGnBu", linewidths=0.5)
plt.title("Heatmap of Students' Scores (Raw Values)")
plt.show()


```
